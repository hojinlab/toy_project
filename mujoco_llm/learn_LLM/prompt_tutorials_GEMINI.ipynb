````python
import os
from dotenv import load_dotenv
from google import genai

load_dotenv()

api_key = os.environ.get("GEMINI_API_KEY")
client = genai.Client(api_key=api_key)
````

````python
# 테스트 실행
try:
    response = client.models.generate_content(
        model="gemini-2.5-flash-lite",
        contents="오늘 저녁 메뉴 추천해줘",
    )
    print(response.text)
except Exception as e:
    print("키 설정이 잘못되었거나 에러가 발생했습니다:", e)
````

````python
# AI에게 주는 업무 지침서. "지침이 있으면 목록으로 만들고, 없으면 없다고 해라"라는 규칙을 정해준다.
sys_instruction = """
당신은 주어진 텍스트를 바탕으로 튜토리얼을 생성하는 AI 어시스턴트입니다.
텍스트에 어떤 절차를 진행하는 방법에 대한 지침이 포함되어 있다면, 
글머리 기호 목록 형식으로 튜토리얼을 생성하십시오.
그렇지 않으면 텍스트에 지침이 포함되어 있지 않음을 사용자에게 알리십시오.
"""
# AI가 처리해야 할 일감.
text_input = """
먼저 평평한 땅을 골라 그라운드 시트를 깝니다. 그 위에 텐트 본체를 펼쳐주세요.
폴대를 조립해서 텐트의 슬리브(구멍)에 X자 모양으로 끼워 넣습니다.
폴대 끝을 텐트 모서리 아일렛(구멍)에 꽂아 텐트를 자립시킵니다.
마지막으로 팩을 45도 각도로 박아서 텐트를 바닥에 고정하고, 
플라이(덮개)를 씌워주면 완성입니다.
하지만 비가 너무 많이 오면 이 텐트는 사용할 수 없습니다.
"""
````

````python
try:
    response = client.models.generate_content(
        model="gemini-2.5-flash-lite", 
        config=genai.types.GenerateContentConfig(
            system_instruction=sys_instruction, # 여기에 시스템 메시지를 넣습니다
        ),
        contents=text_input,
    )
    print("--- 튜토리얼 생성 결과 ---")
    print(response.text)

except Exception as e:
    print(f"에러 발생: {e}")
````

````python
# 사용자 입력 (지침이 없는 엉뚱한 말)
text_input = "오늘은 날씨가 좋네요"

response = client.models.generate_content(
    model="gemini-2.5-flash-lite", 
    config=genai.types.GenerateContentConfig(
        system_instruction=sys_instruction, # 규칙 적용
    ),
    contents=text_input,
)

print("--- 결과 확인 ---")
print(response.text)
````
